# Play count visualizations

#### This notebook is used to get a visual comparison of the plays ran by the top 10 and bottom 10 teams for every season.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def compare_top_bottom(topCSV, bottomCSV):
    # Read df
    top10 = pd.read_csv(topCSV)
    bottom10 = pd.read_csv(bottomCSV)

    # Drop name column
    top10 = top10.drop(columns=["ShortName"])
    bottom10 = bottom10.drop(columns=["ShortName"])

    # Sum values
    topSum = top10.sum(numeric_only=True)
    bottomSum = bottom10.sum(numeric_only=True)

    # Combine 2 df into long format
    combinedDF = pd.DataFrame({
        "PlayType": list(topSum.index) * 2,
        "Count": list(topSum.values) + list(bottomSum.values),
        "Group": ["Top 10"] * len(topSum) + ["Bottom 10"] * len(bottomSum)
    })

    # Plot
    plt.figure(figsize=(12,6))
    sns.barplot(data=combinedDF, x="PlayType", y="Count", hue="Group")
    plt.title("Top 10 vs Bottom 10 Play Counts")
    plt.ylabel("Total Play Counts")
    plt.show()
    return
    
    
```

## 2017 Play Count Visualization

```python
compare_top_bottom("play_counts/2017_top10_play_counts.csv", "play_counts/2017_bottom10_play_counts.csv")
```
